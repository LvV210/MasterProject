```python
import numpy as np
from astropy.time import Time
import pandas as pd
```

#### Make time object with the dates and times like: YYYY-MM-DDTHH:MM:SS.##

```python
spectra_data = pd.read_excel('Spectra/Spectra_Details.xlsx')
falenga = pd.read_excel('tables/FalengaTable1.xlsx').iloc[1:]
```

```python
falenga
```

```python
spectra_data['T(MJD)'] = None
spectra_data['T0(MJD)'] = None
spectra_data['Porb'] = None
# spectra_data['Date'] = pd.to_datetime(spectra_data['Date'])


for index, row in spectra_data.iterrows():
    if str(row['Date']) != 'NaT':
        DateTime = str(row['Date'].strftime('%Y-%m-%d')) + 'T' + str(row['Time'].strftime("%H:%M:%S"))
        spectra_data.loc[index, 'T(MJD)'] = Time(DateTime, format='isot', scale='utc').mjd
        spectra_data.loc[index, 'T0(MJD)'] = falenga.loc[falenga['Source'] == row['OBJECT'], 'Orbital epoch'].reset_index(drop=True).at[0]
        spectra_data.loc[index, 'Porb'] = falenga.loc[falenga['Source'] == row['OBJECT'], 'Orbital period'].reset_index(drop=True).at[0]

phase = (spectra_data['T(MJD)'] - spectra_data['T0(MJD)']) / spectra_data['Porb']
spectra_data['Phase'] = phase - phase[phase.notna()].astype(int)

spectra_data
```
